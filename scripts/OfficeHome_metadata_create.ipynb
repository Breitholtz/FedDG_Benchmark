```python
import torch
import os
from pathlib import Path
import pandas as pd
import numpy as np
```

```python
path = Path("/local/scratch/a/shared/datasets/office_home_v1.0/")
metadata_array = []
OfficeHome_DOMAINS = ['Art', 'Clipart', 'Product', 'Real World']
OfficeHome_CATEGORIES=['Alarm_Clock', 'Backpack', 'Batteries', 'Bed', 'Bike', 'Bottle', 'Bucket', 'Calculator', 'Calendar', 'Candles', 'Chair', 'Clipboards', 'Computer', 'Couch', 'Curtains', 'Desk_Lamp', 'Drill', 'Eraser','Exit_Sign', 'Fan', 'File_Cabinet', 'Flipflops', 'Flowers', 'Folder', 'Fork', 'Glasses', 'Hammer', 'Helmet', 'Kettle', 'Keyboard', 'Knives', 'Lamp_Shade', 'Laptop', 'Marker', 'Monitor', 'Mop', 'Mouse', 'Mug', 'Notebook', 'Oven', 'Pan', 'Paper_Clip', 'Pen', 'Pencil', 'Postit_Notes', 'Printer', 'Push_Pin', 'Radio', 'Refrigerator', 'Ruler', 'Scissors', 'Screwdriver', 'Shelf', 'Sink', 'Sneakers', 'Soda', 'Speaker', 'Spoon', 'Table', 'Telephone', 'ToothBrush', 'Toys', 'Trash_Can', 'TV', 'Webcam']
index = 0
for x in path.iterdir():
    if x.is_dir():
        domain = x.name
        domain_remapped = OfficeHome_DOMAINS.index(domain)
        if domain == 'Art':
            split = 'test'
        elif domain == 'Real World':
            split = 'val'
        else:
            split = 'train'
        for fp in x.iterdir():
            category = fp.name
            y = OfficeHome_CATEGORIES.index(category)
            for im in fp.iterdir():
                image_path = im.relative_to(path)
                metadata_array.append([split, domain_remapped, domain, category, y, str(image_path)])
metadata_np = np.array(metadata_array)
rng = np.random.default_rng()
for dm in ['Clipart', 'Product']:
    for i in range(len(OfficeHome_CATEGORIES)):
        print(i)
        idx = np.logical_and(np.logical_and(metadata_np[:,0]=='train', metadata_np[:,2]==dm), metadata_np[:,4]==str(i)).nonzero()
        val_test = rng.choice(idx[0], np.floor(0.1*len(idx[0])).astype('int'))
        val = val_test[0:len(val_test)//2]
        test = val_test[len(val_test)//2:]
        print(val, test)
        print(len(val_test), len(val_test))
        metadata_np[val, 0] = 'id_val'
        metadata_np[test, 0] = 'id_test'
print(metadata_np)
print(np.unique(metadata_np[:,0]))
pd.DataFrame(metadata_np).to_csv("metadata.csv", index_label=("split", "domain_remapped" "domain", "category", 'y', "path"))
```
